1. Validación Cruzada vs. Predicción Individual
En train.py (74.25%): Estás usando Validación Cruzada (Cross-Validation). Esto significa que el código divide tus 130 mensajes en grupos, entrena con unos y prueba con otros que el modelo nunca ha visto en ese turno. El 74% es una media de estrés; es el peor escenario probable del modelo enfrentándose a datos desconocidos y variados.

En main.py (GUI): Cuando escribes un mensaje, el modelo utiliza el 100% de su conocimiento (entrenado con todos los mensajes del CSV) para darte una respuesta. Al no estar ya en modo "examen" (validación), su confianza estadística sobre un mensaje específico puede ser muy alta.
